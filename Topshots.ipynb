```python
# Define a list of Waterloo players
waterloo_players = [
    "Kymora Stafford", "Mya Jickling", "Alliasen Miscione", "Jaime Newell", 
    "Kate Huh", "Ghiselle Poblete", "Charissa Virr", "Tamara Popovic", 
    "Alyssa Renaud", "Carina Fiore", "Nehita Oko-Oboh", "Anna-Gisele (Gigi) Goulet", 
    "Yemi Oladipo"
]

# Initialize a dictionary to track shot attempts and makes by play type
play_type_stats = {}

# Read the file (assuming the log data is stored in a file called 'vsRegina.txt')
file_path = 'vsUFV.txt'  # Path to the vsRegina file
with open(file_path, 'r') as file:
    game_log = file.readlines()

# Track total shots taken by Waterloo
total_shots = 0

# Process each play from the game log
for play in game_log:
    for player in waterloo_players:
        if player in play:
            try:
                # Extract the play type (the part after the player name and before the result)
                shot_info = play.split(">")[1].strip()
                play_type = shot_info.split(">")[0].strip()

                if play_type not in play_type_stats:
                    play_type_stats[play_type] = {"attempts": 0, "makes": 0}

                # Check if it is a shot attempt (Make or Miss)
                if "Make" in play or "Miss" in play:
                    play_type_stats[play_type]["attempts"] += 1
                    total_shots += 1  # Increment total shots

                    # Check if it is a successful shot (Make)
                    if "Make" in play:
                        play_type_stats[play_type]["makes"] += 1
            except IndexError:
                # Ignore lines that don't follow the expected format
                continue
            break  # Move on once we find the player

# Calculate efficiency (makes / attempts) for each play type
play_type_efficiency_stats = []
for play_type, stats in play_type_stats.items():
    attempts = stats["attempts"]
    makes = stats["makes"]
    if attempts > 0:  # Avoid division by zero
        efficiency = makes / attempts
        play_type_efficiency_stats.append((play_type, efficiency, makes, attempts))

# Sort the play types by efficiency
sorted_play_types = sorted(play_type_efficiency_stats, key=lambda x: x[1], reverse=True)

# Print the total number of shots
print(f"Total shots taken by Waterloo: {total_shots}")

# Print the top 3 play types by efficiency
print("\nTop 3 Play Types by Efficiency:")
for idx, (play_type, efficiency, makes, attempts) in enumerate(sorted_play_types[:5]):
    print(f"{idx+1}. {play_type} - Efficiency: {efficiency:.2f} ({makes} makes / {attempts} attempts)")

# Print the bottom 3 play types by efficiency
print("\nBottom 3 Play Types by Efficiency:")
for idx, (play_type, efficiency, makes, attempts) in enumerate(sorted_play_types[-5:], start=1):
    print(f"{idx}. {play_type} - Efficiency: {efficiency:.2f} ({makes} makes / {attempts} attempts)")
```
